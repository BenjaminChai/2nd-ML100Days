# 作業
- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

%matplotlib inline
```

```python
x = X[:, 1:3]

cls = AgglomerativeClustering(n_clusters=2, linkage="ward")
cls.fit(x)

fig = plt.figure(figsize=(12,5))


ax1 = fig.add_subplot(1,2,1)
ax1.scatter(x[:, 0], x[:, 1], c=y)
ax1.title.set_text('Truth')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x[:, 0], x[:, 1], c=cls.labels_)
ax2.title.set_text('n_clusters=2')
```

```python
x = X[:, 1:3]

cls = AgglomerativeClustering(n_clusters=3, linkage="ward")
cls.fit(x)

fig = plt.figure(figsize=(12,5))


ax1 = fig.add_subplot(1,2,1)
ax1.scatter(x[:, 0], x[:, 1], c=y)
ax1.title.set_text('Truth')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x[:, 0], x[:, 1], c=cls.labels_)
ax2.title.set_text('n_clusters=3')
```
