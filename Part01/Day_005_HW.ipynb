```python
# 檢視AMT_INCOME_TOTAL : 看收入

plt.hist(app_train['AMT_INCOME_TOTAL'], bins=30, log=True)
plt.show()
```

bins要分30，還要用log才能顯示出來，貧富差距也太大了吧 QQ

```python
# 大家都有多少小孩呢？

app_train['CNT_CHILDREN'].value_counts().plot(kind='bar')
plt.show()
```

大部分的人是沒有小孩的，而且小孩數量都是1、2個，3個的已經很少了

```python
# 所以沒小孩的是自己住，還是還在和父母家人住呢

plt.scatter(app_train['CNT_CHILDREN'], app_train['CNT_FAM_MEMBERS'])
plt.xlabel('COUNT CHILDREN')
plt.ylabel('COUNT FAMILY MEMBERS')
plt.xlim(0,20)
plt.ylim(0,20)
plt.show()
```

其實這樣的結果很怪，真的都沒有和家人住的嗎＠＠，都是小孩+1或+2，可以再研究看看。

```python
plt.scatter(app_train['OWN_CAR_AGE'], app_train['AMT_INCOME_TOTAL'])
plt.xlabel('Own Car Age')
plt.ylabel('Amount Income Total')

plt.xlim(0,100)
plt.ylim(0,1000000)
plt.show()
```

- 年輕就有車應該是人生勝利組吧，看起來確實越年輕就有車的整體的收入也較高
- 有一大群人是65歲才第一次有車的，應該是退休買個fun car來玩嗎？
