```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print('資料的row數：', app_train.shape[0])
print('資料的column數：', app_train.shape[1])
```

#### 列出所有欄位

```python
print('列出所有欄位：')
print(app_train.columns)
```

#### 截取部分資料

```python
pd.options.display.max_columns = app_train.shape[1]
print('Top 5 rows:')
app_train.head()
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
print('檢視各位欄的datatype')
app_train.info(verbose=True)
```

```python
print('數值資料列數：', app_train.select_dtypes(['int64', 'float64']).shape[1])
print('類別資料列數：', app_train.select_dtypes('object').shape[1])
```

### 觀察
- 大部分為數值資料
- 有很都被去識別化的欄位需注意

```python
# 檢視null的百分比
na_pct = np.sum(app_train.isnull(), axis=0) / app_train.shape[0]
na_pct[na_pct>0]
```

```python
print('>30% null columns：', na_pct[na_pct>0.3].shape[0])
print('>50% null columns：', na_pct[na_pct>0.5].shape[0])
```

### 觀察
- 41 columns有超過半數的null

```python
pd.options.display.max_columns = app_train.shape[0]
print('檢視數值資料的描述性統計')
app_train.describe()
```

### 觀察
- 有很多columns是描述顧客的boolean flag，裡面若有null值，可能可以用一些產業知識或是眾數來補足，留下True的特徵。
